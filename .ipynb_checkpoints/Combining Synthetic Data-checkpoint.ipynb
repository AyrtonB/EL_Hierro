### Read in and Clean Bootstrap Demand/Wind

```python
s_missing_dts = pd.Series( list(set(dt_rng) - set(pd.to_datetime(df_demand_wind.datetime.unique()))) ).sort_values()
s_missing_dts
```

### Read in and Clean Bootstrap Solar

```python
df_solar = pd.read_csv('data/synth_solar.csv', parse_dates=['datetime'])
df_solar = df_solar.rename(columns={'solar_W':'1_solar_panel'})
df_solar['1_solar_panel'] = df_solar['1_solar_panel']/1000000
print(f'Dataframe size: {df_solar.shape[0]}')
df_solar.head()
```

```python
s_missing_dts = pd.Series( list(set(dt_rng) - set(pd.to_datetime(df_solar.datetime.unique()))) ).sort_values()
s_missing_dts
```

```python
## Creating duplicate values for the 23rd - taken from day before
s_bool_22rd = df_solar.datetime.dt.date == date(2021, 12, 16)
df_23rd = df_solar.loc[s_bool_22rd].copy()
df_23rd.datetime = df_23rd.datetime + timedelta(days=1)

## Dropping 23rd values in df_solar
s_bool_23rd = df_solar.datetime.dt.date == date(2021, 12, 17)
df_solar = df_solar.loc[~s_bool_23rd]

## Inserting new 23rd values
df_solar = df_solar.append(df_23rd)
df_solar = df_solar.sort_values('datetime')
df_solar = df_solar.reset_index(drop=True)

s_missing_dts = pd.Series( list(set(dt_rng) - set(pd.to_datetime(df_solar.datetime.unique()))) ).sort_values()
s_missing_dts
```

```python
print('NaNs')
print( len(df_solar) - df_solar.count() )

print(f'\nShape: {df_solar.shape[0]}')
```

<br>

### Energy System Design

```python
def create_system(df_base_hierro, turbines, panels):
    ## Setting assets
    turbine_size = 2.3 # MW
    panel_size = 4 # W

    current_wind = 5 * turbine_size # MW installed
    current_solar = panel_size # W installed

    new_wind = turbines * turbine_size # MW installed
    new_solar = panels * panel_size # W installed 

    df_base_hierro['5_wind_turbines'] = df_base_hierro['5_wind_turbines'] * (new_wind/current_wind)
    df_base_hierro['1_solar_panel'] = df_base_hierro['1_solar_panel'] * (new_solar/current_solar)

    new_wind_col = f'wind_{new_wind}MW'
    new_solar_col = f'solar_{new_solar/1000}MW'
    df_base_hierro = df_base_hierro.rename(columns={'1_solar_panel':new_solar_col, '5_wind_turbines':new_wind_col})

    df_base_hierro['delta'] = df_base_hierro['demand'] - ( df_base_hierro[new_wind_col]+df_base_hierro[new_solar_col] )

    return df_base_hierro

def calc_imbalance(df_hierro, _print=True):
    gen_shortcoming = df_hierro[df_hierro.delta>=0]['delta'].sum()
    gen_excess = np.abs(df_hierro[df_hierro.delta<0]['delta'].sum())

    if _print == True:
        print(f'Generation shortcoming: {round(gen_shortcoming/(6*1000), 1):,} GWh')
        print(f'Generation excess: {round(gen_excess/(6*1000), 1):,} GWh')
        print('')
        
    return gen_shortcoming, gen_excess

turbines = 5
panels = 2000

df_base_hierro = df_demand_wind.set_index('datetime').copy()
df_base_hierro[df_solar.set_index('datetime').columns] = df_solar.set_index('datetime')
df_base_hierro[df_wind_hmm.set_index('datetime').columns] = df_wind_hmm.set_index('datetime')
df_base_hierro.reset_index().to_csv('data/df_base_hierro.csv', index=False)

df_hierro = hierro.create_hierro(turbines, panels)
gen_shortcoming, gen_excess = hierro.calc_imbalance(df_hierro)
df_hierro.reset_index().to_csv('data/df_hierro_fake.csv', index=False)

df_hierro.head()
```

```python
# Plotting Helper Functions

def plot_sys_dist(df_hierro, turbines, panels, ax=None, label=None, hist=True):
    if ax == None:
        fig = plt.figure()
        ax = plt.subplot()

    sns.distplot(df_hierro['delta'], ax=ax, label=label, hist=hist)

    plt.title(f'{turbines} Turbines, {panels} Panels')
    plt.ylabel('Frequency Density')
    plt.xlabel('Demand Delta')

    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    
    return 

def plot_dem_and_gen(df_hierro):
    fig = plt.figure()
    ax = plt.subplot()

    df_hierro.drop(columns='delta').plot(ax=ax)

    plt.ylabel('Demand/Generation (MW)')
    plt.xlabel('')
    plt.legend(bbox_to_anchor=(1, 1), frameon=False)

    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    
def plot_dem_delta(df_hierro):
    fig = plt.figure()
    ax = plt.subplot()

    df_hierro['delta'].plot(label='Delta')
    pd.Series(np.zeros(df_hierro.shape[0]), df_hierro.index).plot(style='k--')

    plt.ylabel('Demand Delta (MW)')
    plt.xlabel('')

    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
```

```python
min_idx = (1*24*6) * 7*4
max_idx = min_idx + 7*24*6

df_hierro_plot = df_hierro.iloc[min_idx:max_idx].copy()

plot_dem_and_gen(df_hierro_plot)
plot_dem_delta(df_hierro_plot)
```
