# El Hierro Wind Turbine Power Curves

<br>

### Imports

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from scipy.optimize import curve_fit
```

<br>

### Reading in Data

```python
df_gen = pd.read_csv('data/df_hierro_clean.csv', parse_dates=['datetime'])
df_gen.head()
```

```python
df_weather = pd.read_csv('data/weather.csv', parse_dates=['time'])
df_weather.head()
```

```python
wind_cap_factor = (df_gen[df_gen.datetime.dt.year==2018].wind/(2.3*5)).mean()
wind_cap_factor
```

<br>

### Creating Wind Dataframe

```python
s_bool_dup_idx = df_weather.set_index('time').index.duplicated()

num_dup_idxs = s_bool_dup_idx.sum()
print(f'Number of duplicate dates in index: {num_dup_idxs}')

df_weather = df_weather[~s_bool_dup_idx]
```

```python
df_wind = pd.DataFrame()

df_wind['gen'] = df_gen.set_index('datetime').resample('30T').mean()['wind']
df_wind['direction'] = df_weather.set_index('time')['wdir']
df_wind['speed'] = df_weather.set_index('time')['wspd']

df_wind = df_wind.dropna()
df_wind.head()
```

<br>

### Visualising Missing Datetimes

```python
min_date = df_wind.index.min()
max_date = df_wind.index.max()

print(f'Earliest Date: {min_date}')
print(f'Latest Date: {max_date}')
```

```python
s_dt_idx = pd.date_range(min_date, max_date, freq='30T')
s_dt_rng = pd.Series(index=s_dt_idx)
s_dt_rng.loc[:] = 0

s_dt_missing = pd.Series(list( set(s_dt_idx.values) - set(df_wind.index.values) ))
s_dt_missing = s_dt_missing.sort_values().reset_index(drop=True)
s_dt_rng.loc[s_dt_missing] = 1

print(s_dt_missing.shape)
s_dt_missing.head()
```

```python
## Create dataframe
df_dt_rng = pd.DataFrame(s_dt_rng).rename(columns={0:'NaN'})
df_dt_rng = df_dt_rng.resample('60T').sum()

## Separating date and time
df_dt_rng['time'] = df_dt_rng.index.time
df_dt_rng['date'] = df_dt_rng.index.date

## Pivoting to wide format
df_dt_rng = df_dt_rng.pivot('date', 'time', 'NaN')

## Resampling to monthly 
df_dt_rng.index = pd.to_datetime(df_dt_rng.index)
df_dt_rng = df_dt_rng.resample('M').sum()
df_dt_rng = 100*df_dt_rng/30

df_dt_rng.head()
```

```python
## Plotting
fig = plt.figure(dpi=150)
ax = plt.subplot()

sns.heatmap(df_dt_rng, ax=ax, cmap='Blues', cbar_kws={'label': 'Percentage Missing Values'})

xlabels = [item for item in ax.get_xticklabels()]
ax.set_xticklabels([label.get_text()[:2] for label in xlabels])

ylabels = [item for item in ax.get_yticklabels()]
ax.set_yticklabels([pd.to_datetime(label.get_text()[:10]).strftime('%B %Y') for label in ylabels])

plt.xlabel('Time')
plt.ylabel('')
```

<br>

### Plotting Distribution of Wind Speed & Generation

### Power Curve

<img src='img/power_curve.PNG' width=500 class="center">

```python
df_power_curve = (pd
                  .read_csv('data/power_curve.csv', header=None)
                  .rename(columns={0:'wind_speed_ms', 1:'power_kW'})
                 )

df_power_curve.loc[df_power_curve.power_kW<0, 'power_kW'] = 0
df_power_curve = df_power_curve[df_power_curve.power_kW<3000]
df_power_curve.head()
```

```python
mph = df_power_curve.wind_speed_ms
power = df_power_curve.power_kW*5/1000

plt.plot(mph, power)
```

<br>

Here we fit the power curve to the observed data using a transfer on both power and speed. We assume effective wind speed is 4.5 m/s lower than measured, and that power is curtailed at levels of around 25%.

```python
##### Plotting
fig = plt.figure(dpi=1000)
ax = plt.subplot()

sns.kdeplot(df_wind.speed, df_wind.gen, ax=ax, shade=True, cmap='Blues', label='Frequency Density')
plt.plot((mph), power*0.75, 'k--', label='Adjusted Power Curve')

plt.xlim(0, 30)
plt.ylim(0, 12)

plt.xlabel('Wind Speed (mph)')
plt.ylabel('Total Generation (MW)')
#plt.title('El Hierro 5 Turbine Array')

plt.legend(frameon=False, loc=2)

#ax.spines['right'].set_visible(False)
#ax.spines['top'].set_visible(False)

plt.savefig('img/wind_curve_actual.png', dpi=1000, bbox_inches='tight')
```
