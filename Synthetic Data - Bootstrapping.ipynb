### Reading in Data & Basic Checks

```python
import pandas as pd
import numpy as np

from datetime import datetime, time, date

import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns

from hmmlearn.hmm import GaussianHMM

import random
```

```python
df_hierro = pd.read_csv('data/el_hierro_16_18.csv', parse_dates=['datetime'])
df_hierro.head()
```

```python
## Assigning colours to months
month_colours = sns.color_palette('viridis', 12)
sns.palplot(month_colours)

months = range(1, 13)
month_names = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

months_palette = dict(zip(months, month_colours))
```

```python
## Plotting Helper Functions
def plot_avg_day(df, ax, datetime_col='datetime', param_col='demand', label=None, color=None):
    ## Given a timeseries
    s_times = df[datetime_col].dt.time
    
    s_avg_day = (df
                 .groupby(s_times)
                 .mean()
                 [param_col]
                )

    datetimes_ord = times_2_ord(s_avg_day.index.values)
    ax.plot(datetimes_ord, s_avg_day, label=label, color=color)
    
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    
def times_2_ord(times):
    ## Takes a datetime.time array and converts it to ordinal data Matplotlib can handle
    today = date.today()
    datetimes = [ datetime.combine(today, time) for time in times ]
    datetimes_ord = mdates.date2num(datetimes)
    return datetimes_ord


###################################################################
## Creating Combined Plot of Average Daily Load Profile by Month ##
###################################################################

fig = plt.figure(dpi=150)
ax = plt.subplot()

for month in months:
    df_hierro_month = df_hierro[(df_hierro.datetime.dt.month==month)&(df_hierro.datetime.dt.year==2016)]
    
    plot_avg_day(df_hierro_month, ax, label=month_names[month-1], color=months_palette[month])
    
ax.set_title('Daily Load Profile by Month')
ax.legend(title='Month', frameon=False, bbox_to_anchor=(1, 1))

ax.xaxis.set_major_locator(mdates.HourLocator(interval=4))    
ax.xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))

times = ax.lines[0].get_xdata()
ax.set_xlim(times[0], round(times[-1]))
```

### Synthetic Time Series of Demand through Bootstrapping

```python
def generate_fake_datetime(df_hierro_month_day, year, month, day):
    s_time = df_hierro_month_day.datetime.dt.time
    df_datetime = pd.DataFrame({'time':s_time})
    _date = date(year, month, day)
    df_datetime['date'] = _date

    df_datetime['datetime'] = pd.to_datetime(df_datetime['date'].astype(str) + ' ' + df_datetime['time'].astype(str))
    df_hierro_month_day = df_hierro_month_day.assign(datetime=df_datetime['datetime'])
    
    return df_hierro_month_day

def create_single_month_synth(df_hierro_month, uniq_dts, year, month, month_length):
    uniq_dts_for_month = uniq_dts.shape[0]

    rnd_dts = [] # These are the random dates selected during bootstrapping
    df_hierro_month_days = pd.DataFrame()

    # Randomly selecting dates via bootstrapping d creating synthetic time series
    for i in range(month_length):
        rnd_dt = uniq_dts[random.randint(0, uniq_dts_for_month-1)]
        rnd_dts.append(rnd_dt)
        
        df_hierro_month_day = df_hierro_month[df_hierro_month.datetime.dt.date==rnd_dt]
        
        df_hierro_month_day = generate_fake_datetime(df_hierro_month_day, year, month, i+1)
        df_hierro_month_days = df_hierro_month_days.append(df_hierro_month_day)
        
    return df_hierro_month_days
```

```python
month_lengths = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
month_2_month_length = dict(zip(months, month_lengths))
year = 2021

df_hierro_year = pd.DataFrame()

for month in months:
    df_hierro_month = df_hierro[df_hierro.datetime.dt.month==month]
    
    uniq_dts = df_hierro_month.datetime.dt.date.unique()
    month_length = month_2_month_length[month]
    
    df_hierro_month = create_single_month_synth(df_hierro_month, uniq_dts, year, month, month_length)
    df_hierro_year = df_hierro_year.append(df_hierro_month)
    
df_hierro_year.head()
```

```python
fig = plt.figure(dpi=150)
ax = plt.subplot()

for month in months:
    df_hierro_month = df_hierro_year[df_hierro_year.datetime.dt.month==month]
    
    plot_avg_day(df_hierro_month, ax, label=month_names[month-1], color=months_palette[month])
    
ax.set_title('Daily Load Profile by Month')
ax.legend(title='Month', frameon=False, bbox_to_anchor=(1, 1))

ax.xaxis.set_major_locator(mdates.HourLocator(interval=4))    
ax.xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))

times = ax.lines[0].get_xdata()
ax.set_xlim(times[0], round(times[-1]))
```
